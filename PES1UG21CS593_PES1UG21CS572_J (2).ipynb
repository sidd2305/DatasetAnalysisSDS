```python
#Linear regression model for active and rest-evaluating goodness of fit
from sklearn.metrics import r2_score
x=np.array(df["Rest"].tolist())
y=np.array(df["Active"].tolist())
print("Rest",x)
print("Active",y)
plt.plot(x,y,"ro")
b1,b0=np.polyfit(x,y,deg=1)
print("y hat = ",b0,"+",b1,"x")
ypred=b0+b1*x
plt.plot(x,ypred)
plt.show()
print("R square is",r2_score(y,ypred))
residual=y-ypred
plt.plot(ypred,residual,"ro")
plt.hlines(0,0,12000)
#Linear regression is a fair representation model for the given data
```

```python
#Fitted model equation and FINDING R squared value
from sklearn.metrics import r2_score
x=df.iloc[:,1]
y=df.iloc[:,2]

plt.plot(x,y,'ro')
b1,b0=np.polyfit(x,y,deg=1)
print("Fitted model linear equation","y_hat=",b0,"+",b1,"x")
ypredicted=b0+b1*x#predcited or fitted value
plt.plot(x,ypredicted)
plt.show()
rsq=r2_score(y,ypredicted)
print("R squared value",rsq)
```
